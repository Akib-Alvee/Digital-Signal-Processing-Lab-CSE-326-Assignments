**Unit Impulse**

For unit impulse signal, $\begin{equation}
    \delta(n)=
    \begin{cases}
      1, & \text{for}\ n=0 \\
      0, & \text{for}\ n \neq 0
    \end{cases}
  \end{equation}$

```python
def Delay(x_n,k):
  ret=np.zeros(abs(k))

  if(k>=0):
    ret=np.concatenate((ret,x_n[0:len(x_n)-k]))
  else:
    ret=np.concatenate((x_n[abs(k):len(x_n)],ret))
  return ret
```

```python
lim = 60
n=np.linspace(-lim,lim,61)
x_n=np.zeros(61)

p=np.where(n==0)
x_n[p]=1

fig = plt.figure(figsize = (20, 8))
fig.suptitle('Unit Impulse Signal', fontsize=18)

plt.subplot(1,2,1)
plt.title('Without Time Shifting',fontsize=14)
plt.stem(n, x_n)
plt.axvline(linewidth='1', color='r')
plt.axhline(linewidth='1', color='r')
k=-10
delayed_signal=Delay(x_n,k)
plt.subplot(1,2,2)
plt.title('With Time Shifting',fontsize=14)
plt.stem(n, delayed_signal)
plt.axvline(linewidth='1', color='r')
plt.axhline(linewidth='1', color='r')
```

**Unit Step**

For unit step signal, $\begin{equation}
    u(n)=
    \begin{cases}
      1, & \text{for}\ n\geq0 \\
      0, & \text{for}\ n < 0
    \end{cases}
  \end{equation}$

```python
lim = 60
n=np.linspace(-lim,lim,61)
x_n=np.zeros(61)

p=np.where(n>=0)
x_n[p]=1

fig = plt.figure(figsize = (20, 8))
fig.suptitle('Unit Step Signal', fontsize=18)

plt.subplot(1,2,1)
plt.title('Without Time Shifting',fontsize=14)
plt.stem(n, x_n)
plt.axvline(linewidth='1', color='r')
plt.axhline(linewidth='1', color='r')

k=10
delayed_signal=Delay(x_n,k)
plt.subplot(1,2,2)
plt.title('With Time Shifting',fontsize=14)
plt.stem(n, delayed_signal)
plt.axvline(linewidth='1', color='r')
plt.axhline(linewidth='1', color='r')
```

**Unit Ramp**

For unit ramp signal, $\begin{equation}
    u_r(n)=
    \begin{cases}
      n, & \text{for}\ n\geq0 \\
      0, & \text{for}\ n < 0
    \end{cases}
  \end{equation}$

```python
lim = 60
n=np.linspace(-lim,lim,61)
x_n=np.zeros(61)

p=np.where(n>=0)
x_n[p]=1

fig = plt.figure(figsize = (20, 8))
fig.suptitle('Unit Ramp Signal', fontsize=18)

plt.subplot(1,2,1)
plt.title('Without Time Shifting',fontsize=14)
plt.stem(n, x_n*n)
plt.axvline(linewidth='1', color='r')
plt.axhline(linewidth='1', color='r')

k=10
delayed_signal=Delay(x_n,k)
plt.subplot(1,2,2)
plt.title('With Time Shifting',fontsize=14)
plt.stem(n, delayed_signal*n)
plt.axvline(linewidth='1', color='r')
plt.axhline(linewidth='1', color='r')
```

**Exponential Signal**

For exponential signal, $ x(n)=a^n\hspace{5pt} \text{for all n}$

```python
n = np.arange(0,100,1)
def exponentialSignal(a):
  y = []

  for i in range(len(n)):
      temp = a**n[i]
      y.append(temp)
  return y
    

# #Plotting the graph

fig = plt.figure(figsize = (10,20))
fig.suptitle('Exponential Signal', fontsize=30)

plt.subplot(4,1,1)
plt.title('0 < a < 1',fontsize=14)
plt.stem(n,exponentialSignal(.5), 'r')
plt.axvline(linewidth='1', color='r')
plt.axhline(linewidth='1', color='r')
fig.tight_layout(pad=2.0)

plt.subplot(4,1,2)
plt.title(' a > 1',fontsize=14)
plt.stem(n,exponentialSignal(1.5), 'r')
plt.axvline(linewidth='1', color='r')
plt.axhline(linewidth='1', color='r')
fig.tight_layout(pad=2.0)

plt.subplot(4,1,3)
plt.title('-1 < a < 0',fontsize=14)
plt.stem(n,exponentialSignal(-.5), 'r')
plt.axvline(linewidth='1', color='r')
plt.axhline(linewidth='1', color='r')
fig.tight_layout(pad=2.0)

plt.subplot(4,1,4)
plt.title(' a < -1',fontsize=14)
plt.stem(n,exponentialSignal(-1.5), 'r')
plt.axvline(linewidth='1', color='r')
plt.axhline(linewidth='1', color='r')
fig.tight_layout(pad=2.0)
```
