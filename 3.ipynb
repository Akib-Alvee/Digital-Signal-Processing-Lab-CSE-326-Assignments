# 3. Compute and plot Discrete Fourier Transform (DFT) of a signal using DFT equation.

```python
# import necessary library

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
import cmath
```

```python
from IPython.core.display import HTML
HTML("""
<style>
.output_png {
    display: table-cell;
    text-align: center;
    vertical-align: middle;
}
</style>
""")
```

```python
n = np.arange(0, 100, 1)
y = np.zeros(len(n))

for i in range(len(n)):
    y[i] = np.random.randint(-5,5,dtype = int)

def DFT(y, N):
    temp = 2*np.pi / N 
    X=np.zeros(N,dtype=np.complex)
    X_K = np.zeros(N)
    for K in range(N):
        for n in range(N):
            X[K] = X[K]+y[n]*np.exp(-1j*K*temp*n)
    
    return X

x_k = np.zeros(len(y))
x_k = DFT(y,len(y))

magnitude = np.absolute(x_k)
phase = np.angle(x_k)

fig = plt.figure(figsize = (20,15))
fig.suptitle('Discrete Fourier Transform (DFT) ', fontsize=30)
n = np.arange(0,len(y),1)

plt.subplot(2,2,1)
plt.title('Input Signal',fontsize=20)
plt.stem(n,y,'r')
plt.axvline(linewidth='1', color='r')
plt.axhline(linewidth='1', color='r')
fig.tight_layout(pad=2.0)

plt.subplot(2,2,2)
plt.title('Magnitude',fontsize=20)
plt.stem(n,magnitude,'r')
plt.axvline(linewidth='1', color='r')
plt.axhline(linewidth='1', color='r')
fig.tight_layout(pad=2.0)

plt.subplot(2,2,3)
plt.title('Phase',fontsize=20)
plt.stem(n,phase,'r')
plt.axvline(linewidth='1', color='r')
plt.axhline(linewidth='1', color='r')
fig.tight_layout(pad=2.0)
```
