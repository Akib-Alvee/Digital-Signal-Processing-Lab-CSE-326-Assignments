```python
combofile = "sound_with_noise.wav"
staticfile= "noise.wav"

FsCombo,comboData=read(combofile)
FsStatic,staticData=read(staticfile)
print("Sampling Frequency",FsCombo)

plt.figure(figsize=(15,10))
plt.ylim(-3000,3000)
plt.xlim(0,2500)
plt.plot(comboData)
plt.plot(staticData)
# w = wave.open(file, 'r')
# for i in range(w.getnframes()):
#     frame = w.readframes(i)
#     print(frame)
```

Apply DFT

```python

fft_combo= fft(comboData)
fft_static=fft(staticData)

clear=fft_combo-fft_static

print(fft_combo)
print(fft_static)
print(clear)
```

Inverse FFT

```python

filtered_sound=ifft(clear)
plt.figure(figsize=(15,10))
plt.ylim(-3000,3000)
plt.xlim(0,2500)
plt.plot(filtered_sound)
```

```python

sound_file=write("filter.wav",FsCombo,filtered_sound.astype(comboData.dtype))
```
