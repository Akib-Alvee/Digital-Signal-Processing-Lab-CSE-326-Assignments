# 4. Compute the inverse DFT of the signal obtained in (3).

```python
# import necessary library

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
import cmath
```

```python
n = np.arange(0, 100, 1)
x = np.zeros(len(n))

for i in range(len(n)):
    x[i] = np.random.randint(-5,5,dtype = int)

def DFT(x, N):
    temp = 2*np.pi / N 
    X=np.zeros(N,dtype=np.complex)
    for K in range(N):
        for n in range(N):
            X[K] = X[K]+x[n]*np.exp(-1j*K*temp*n)
    
    return X

def IDFT(x,N):
    temp = 2*np.pi / N 
    x_n = np.zeros(N)
    for n in range(N):
        for K in range(N):
            x_n[n] = np.real(x_n[n]+x[K]*np.exp(1j*K*temp*n))
        x_n[n] = x_n[n]/N
#     x_n = np.real(x_n)
    x_n = np.round(x_n,decimals=0)
    return x_n

x_k = np.zeros(len(x))
x_k = DFT(x,len(x))

magnitude = np.absolute(x_k)
phase = np.angle(x_k)

fig = plt.figure(figsize = (20,15))
fig.suptitle('Discrete Fourier Transform (DFT) ', fontsize=30)
n = np.arange(0,len(x),1)

plt.subplot(2,2,1)
plt.title('Input Signal',fontsize=20)
plt.stem(n,x,'r')
plt.axvline(linewidth='1', color='r')
plt.axhline(linewidth='1', color='r')
fig.tight_layout(pad=2.0)

plt.subplot(2,2,2)
plt.title('Magnitude',fontsize=20)
plt.stem(n,magnitude,'r')
plt.axvline(linewidth='1', color='r')
plt.axhline(linewidth='1', color='r')
fig.tight_layout(pad=2.0)

plt.subplot(2,2,3)
plt.title('Phase',fontsize=20)
plt.stem(n,phase,'r')
plt.axvline(linewidth='1', color='r')
plt.axhline(linewidth='1', color='r')
fig.tight_layout(pad=2.0)

plt.subplot(2,2,4)
plt.title('Inverse DFT',fontsize=20)
plt.stem(n,IDFT(x_k,len(x)),'r')
plt.axvline(linewidth='1', color='r')
plt.axhline(linewidth='1', color='r')
fig.tight_layout(pad=2.0)
```
