# 2. Compute linear convolution of two signals. Compare the result with the output of the Matlab/Python library function.

```python
# Importing necessary libraries

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
import numpy as np

numberOfValues_x = 10
numberOfValues_h = 12

x = np.zeros(numberOfValues_x)
h = np.zeros(numberOfValues_h)

for i in range(numberOfValues_x):
    x[i] = np.random.randint(-5,5,dtype = int)

for i in range(numberOfValues_h):
    h[i] = np.random.randint(-5,5,dtype = int)

y = np.convolve(x,h)

len_x = len(x)
len_h = len(h)
n = len_x+len_h-1

additionalZeros_x = n - len_x
additionalZeros_h = n - len_h

x_prev = x
h_prev = h

x = np.pad(x,(0,additionalZeros_x),'constant')
h = np.pad(h,(0,additionalZeros_h),'constant')

result = np.zeros(n)

for i in range(n):
    for j in range(n):
        if i>=j:
            result[i] = result[i] + (x[j] * h[i-j])

print ('Linear Convolution by computation :\n',result)

print('Linear Convolution by Library Function : \n',y)
```

```python
fig = plt.figure(figsize = (20, 30))
fig.suptitle('Linear Convolution', fontsize=30)
n = np.arange(0,21,1)


plt.subplot(4,1,1)
plt.title('Input Signal',fontsize=20)
plt.stem(n,x)
plt.axvline(linewidth='1', color='r')
plt.axhline(linewidth='1', color='r')
fig.tight_layout(pad=2.0)

plt.subplot(4,1,2)
plt.title('Impulse Response',fontsize=20)
plt.stem(n,h)
plt.axvline(linewidth='1', color='r')
plt.axhline(linewidth='1', color='r')
fig.tight_layout(pad=2.0)

plt.subplot(4,1,3)
plt.title('Linear Convolution by computation',fontsize=20)
plt.stem(n,result)
plt.axvline(linewidth='1', color='r')
plt.axhline(linewidth='1', color='r')
fig.tight_layout(pad=2.0)

plt.subplot(4,1,4)
plt.title('Linear Convolution using Library Function',fontsize=20)
plt.stem(n,y)
plt.axvline(linewidth='1', color='r')
plt.axhline(linewidth='1', color='r')
fig.tight_layout(pad=2.0)
```
